```python
import csv
import math
import matplotlib.pyplot as plt

ceresTrue = './outlier10%/ceresTrue.csv'
ceresNull = './outlier10%/ceresNull.csv'
ceresCauchy = './outlier10%/ceresCauchy2.0.csv'
ceresHuber = './outlier10%/ceresHuber8.0.csv'
ceresTukey = './outlier10%/ceresTukey8.0.csv'
ceresArctan = './outlier10%/ceresArctan10.0.csv'

trueXji = []
trueYji = []
trueThetaji = []
nullXji = []
nullYji = []
nullThetaji = []
cauchyXji = []
cauchyYji = []
cauchyThetaji = []
huberXji = []
huberYji = []
huberThetaji = []
tukeyXji = []
tukeyYji = []
tukeyThetaji = []
arctanXji = []
arctanYji = []
arctanThetaji = []

def calculate_rmse(y_true, y_pred):
    if len(y_true) != len(y_pred):
        raise ValueError("Input lists must have the same length")

    squared_erros = [(true - pred) ** 2 for true, pred in zip(y_true, y_pred)]
    mean_squared_error = sum(squared_erros) / len(y_true)
    rmse = math.sqrt(mean_squared_error)
    return rmse

# Read data from CSV file
with open(ceresTrue, 'r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader) # Skip the header row if there is one

    for row in csv_reader:
        trueXji.append(float(row[0]))
        trueYji.append(float(row[1]))
        trueThetaji.append(float(row[2]))

with open(ceresNull, 'r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader) # Skip the header row if there is one

    for row in csv_reader:
        nullXji.append(float(row[0]))
        nullYji.append(float(row[1]))
        nullThetaji.append(float(row[2]))

with open(ceresCauchy, 'r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader) # Skip the header row if there is one

    for row in csv_reader:
        cauchyXji.append(float(row[0]))
        cauchyYji.append(float(row[1]))
        cauchyThetaji.append(float(row[2]))

with open(ceresHuber, 'r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader) # Skip the header row if there is one

    for row in csv_reader:
        huberXji.append(float(row[0]))
        huberYji.append(float(row[1]))
        huberThetaji.append(float(row[2]))

with open(ceresTukey, 'r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader) # Skip the header row if there is one

    for row in csv_reader:
        tukeyXji.append(float(row[0]))
        tukeyYji.append(float(row[1]))
        tukeyThetaji.append(float(row[2]))

with open(ceresArctan, 'r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader) # Skip the header row if there is one

    for row in csv_reader:
        arctanXji.append(float(row[0]))
        arctanYji.append(float(row[1]))
        arctanThetaji.append(float(row[2]))

rmseNullX = calculate_rmse(trueXji, nullXji)
rmseNullY = calculate_rmse(trueYji, nullYji)
rmseNullT = calculate_rmse(trueThetaji, nullThetaji)

rmseCauchyX = calculate_rmse(trueXji, cauchyXji)
rmseCauchyY = calculate_rmse(trueYji, cauchyYji)
rmseCauchyT = calculate_rmse(trueThetaji, cauchyThetaji)

rmseHuberX = calculate_rmse(trueXji, huberXji)
rmseHuberY = calculate_rmse(trueYji, huberYji)
rmseHuberT = calculate_rmse(trueThetaji, huberThetaji)

rmseTukeyX = calculate_rmse(trueXji, tukeyXji)
rmseTukeyY = calculate_rmse(trueYji, tukeyYji)
rmseTukeyT = calculate_rmse(trueThetaji, tukeyThetaji)

rmseArctanX = calculate_rmse(trueXji, arctanXji)
rmseArctanY = calculate_rmse(trueYji, arctanYji)
rmseArctanT = calculate_rmse(trueThetaji, arctanThetaji)

plt.subplot(3,1,1)
plt.suptitle('mEstimators with 10% outlier') 
plt.title('$X_{ji}$')
plt.plot(trueXji, label='true')
plt.plot(nullXji, label='l2-norm')
plt.plot(cauchyXji, label='cauchy')
plt.plot(huberXji, label='huber')
plt.plot(tukeyXji, label='tukey')
plt.plot(arctanXji, label='arctan')
plt.legend()
plt.grid()

plt.subplot(3,1,2)
plt.title('$Y_{ji}$')
plt.plot(trueYji, label='true')
plt.plot(nullYji, label='l2-norm')
plt.plot(cauchyYji, label='cauchy')
plt.plot(huberYji, label='huber')
plt.plot(tukeyYji, label='tukey')
plt.plot(arctanYji, label='arctan')
plt.grid()

plt.subplot(3,1,3)
plt.title('$\Theta_{ji}$')
plt.plot(trueThetaji, label='true')
plt.plot(nullThetaji, label='l2-norm')
plt.plot(cauchyThetaji, label='cauchy')
plt.plot(huberThetaji, label='huber')
plt.plot(tukeyThetaji, label='tukey')
plt.plot(arctanThetaji, label='arctan')
plt.grid()
plt.savefig('./outlier10%/mEstimator_10%.png', bbox_inches='tight')
plt.show()


labels = ['L2-norm', 'Cauchy', 'Huber', 'Tukey', 'Arctan']
values = [rmseNullX, rmseCauchyX, rmseHuberX, rmseTukeyX, rmseArctanX]

plt.bar(labels, values, color='green')
plt.title('RMSE with 10% outliers')
plt.grid(True, axis='y')
plt.savefig('./outlier10%/rmse_10%.png', bbox_inches='tight')
plt.show()
```

```python
import csv
import math
import matplotlib.pyplot as plt

ceresTrue = './outlier50%/ceresTrue.csv'
ceresNull = './outlier50%/ceresNull.csv'
ceresCauchy = './outlier50%/ceresCauchy20.0.csv'
ceresHuber = './outlier50%/ceresHuber10.0.csv'
ceresTukey = './outlier50%/ceresTukey8.0.csv'
ceresArctan = './outlier50%/ceresArctan10.0.csv'

trueXji = []
trueYji = []
trueThetaji = []
nullXji = []
nullYji = []
nullThetaji = []
cauchyXji = []
cauchyYji = []
cauchyThetaji = []
huberXji = []
huberYji = []
huberThetaji = []
tukeyXji = []
tukeyYji = []
tukeyThetaji = []
arctanXji = []
arctanYji = []
arctanThetaji = []

def calculate_rmse(y_true, y_pred):
    if len(y_true) != len(y_pred):
        raise ValueError("Input lists must have the same length")

    squared_erros = [(true - pred) ** 2 for true, pred in zip(y_true, y_pred)]
    mean_squared_error = sum(squared_erros) / len(y_true)
    rmse = math.sqrt(mean_squared_error)
    return rmse

# Read data from CSV file
with open(ceresTrue, 'r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader) # Skip the header row if there is one

    for row in csv_reader:
        trueXji.append(float(row[0]))
        trueYji.append(float(row[1]))
        trueThetaji.append(float(row[2]))

with open(ceresNull, 'r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader) # Skip the header row if there is one

    for row in csv_reader:
        nullXji.append(float(row[0]))
        nullYji.append(float(row[1]))
        nullThetaji.append(float(row[2]))

with open(ceresCauchy, 'r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader) # Skip the header row if there is one

    for row in csv_reader:
        cauchyXji.append(float(row[0]))
        cauchyYji.append(float(row[1]))
        cauchyThetaji.append(float(row[2]))

with open(ceresHuber, 'r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader) # Skip the header row if there is one

    for row in csv_reader:
        huberXji.append(float(row[0]))
        huberYji.append(float(row[1]))
        huberThetaji.append(float(row[2]))

with open(ceresTukey, 'r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader) # Skip the header row if there is one

    for row in csv_reader:
        tukeyXji.append(float(row[0]))
        tukeyYji.append(float(row[1]))
        tukeyThetaji.append(float(row[2]))

with open(ceresArctan, 'r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader) # Skip the header row if there is one

    for row in csv_reader:
        arctanXji.append(float(row[0]))
        arctanYji.append(float(row[1]))
        arctanThetaji.append(float(row[2]))

rmseNullX = calculate_rmse(trueXji, nullXji)
rmseNullY = calculate_rmse(trueYji, nullYji)
rmseNullT = calculate_rmse(trueThetaji, nullThetaji)

rmseCauchyX = calculate_rmse(trueXji, cauchyXji)
rmseCauchyY = calculate_rmse(trueYji, cauchyYji)
rmseCauchyT = calculate_rmse(trueThetaji, cauchyThetaji)

rmseHuberX = calculate_rmse(trueXji, huberXji)
rmseHuberY = calculate_rmse(trueYji, huberYji)
rmseHuberT = calculate_rmse(trueThetaji, huberThetaji)

rmseTukeyX = calculate_rmse(trueXji, tukeyXji)
rmseTukeyY = calculate_rmse(trueYji, tukeyYji)
rmseTukeyT = calculate_rmse(trueThetaji, tukeyThetaji)

rmseArctanX = calculate_rmse(trueXji, arctanXji)
rmseArctanY = calculate_rmse(trueYji, arctanYji)
rmseArctanT = calculate_rmse(trueThetaji, arctanThetaji)

plt.subplot(3,1,1)
plt.suptitle('mEstimators with 50% outlier') 
plt.title('$X_{ji}$')
plt.plot(trueXji, label='true')
plt.plot(nullXji, label='l2-norm')
plt.plot(cauchyXji, label='cauchy')
plt.plot(huberXji, label='huber')
plt.plot(tukeyXji, label='tukey')
plt.plot(arctanXji, label='arctan')
plt.legend()
plt.grid()

plt.subplot(3,1,2)
plt.title('$Y_{ji}$')
plt.plot(trueYji, label='true')
plt.plot(nullYji, label='l2-norm')
plt.plot(cauchyYji, label='cauchy')
plt.plot(huberYji, label='huber')
plt.plot(tukeyYji, label='tukey')
plt.plot(arctanYji, label='arctan')
plt.grid()

plt.subplot(3,1,3)
plt.title('$\Theta_{ji}$')
plt.plot(trueThetaji, label='true')
plt.plot(nullThetaji, label='l2-norm')
plt.plot(cauchyThetaji, label='cauchy')
plt.plot(huberThetaji, label='huber')
plt.plot(tukeyThetaji, label='tukey')
plt.plot(arctanThetaji, label='arctan')
plt.grid()
plt.savefig('./outlier50%/mEstimator_50%.png', bbox_inches='tight')
plt.show()


labels = ['L2-norm', 'Cauchy', 'Huber', 'Tukey', 'Arctan']
values = [rmseNullX, rmseCauchyX, rmseHuberX, rmseTukeyX, rmseArctanX]

plt.bar(labels, values, color='green')
plt.title('RMSE with 50% outliers')
plt.grid(True, axis='y')
plt.savefig('./outlier50%/rmse_50%.png', bbox_inches='tight')
plt.show()
```
