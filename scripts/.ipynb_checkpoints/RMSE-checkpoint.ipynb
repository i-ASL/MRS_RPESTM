```python
from sklearn.metrics import mean_squared_error

## RMSE of KF
RMSE_kf_case1_xji = mean_squared_error(cat_true_x_ji_list, cat_one_x_ji_list)**0.5
RMSE_kf_case1_yji = mean_squared_error(cat_true_y_ji_list, cat_one_y_ji_list)**0.5
RMSE_kf_case1_thetaji = mean_squared_error(cat_true_theta_ji_list, cat_one_theta_ji_list)**0.5
RMSE_kf_case1_total = (RMSE_kf_case1_xji + RMSE_kf_case1_yji+ RMSE_kf_case1_thetaji) / 3
print('The x,y,theta RMSE of case 1 :', RMSE_kf_case1_xji, RMSE_kf_case1_yji, RMSE_kf_case1_thetaji)

RMSE_kf_case2_xji = mean_squared_error(cat_true_x_ji_list, cat_two_x_ji_list)**0.5
RMSE_kf_case2_yji = mean_squared_error(cat_true_y_ji_list, cat_two_y_ji_list)**0.5
RMSE_kf_case2_thetaji = mean_squared_error(cat_true_theta_ji_list, cat_two_theta_ji_list)**0.5
RMSE_kf_case2_total = (RMSE_kf_case2_xji + RMSE_kf_case2_yji + RMSE_kf_case2_thetaji) / 3
print('The x,y,theta RMSE of case 2 :', RMSE_kf_case2_xji, RMSE_kf_case2_yji, RMSE_kf_case2_thetaji)

RMSE_kf_case3_xji = mean_squared_error(cat_true_x_ji_list, cat_three_x_ji_list)**0.5
RMSE_kf_case3_yji = mean_squared_error(cat_true_y_ji_list, cat_three_y_ji_list)**0.5
RMSE_kf_case3_thetaji = mean_squared_error(cat_true_theta_ji_list, cat_three_theta_ji_list)**0.5
RMSE_kf_case3_total = (RMSE_kf_case3_xji + RMSE_kf_case3_yji + RMSE_kf_case3_thetaji) / 3
print('The x,y,theta RMSE of case 3 :', RMSE_kf_case3_xji, RMSE_kf_case3_yji, RMSE_kf_case3_thetaji)

RMSE_kf_case4_xji = mean_squared_error(cat_true_x_ji_list, cat_four_x_ji_list)**0.5
RMSE_kf_case4_yji = mean_squared_error(cat_true_y_ji_list, cat_four_y_ji_list)**0.5
RMSE_kf_case4_thetaji = mean_squared_error(cat_true_theta_ji_list, cat_four_theta_ji_list)**0.5
RMSE_kf_case4_vji = mean_squared_error(trueVj, cat_four_v_j_list)**0.5
RMSE_kf_case4_wji = mean_squared_error(trueWj, cat_four_w_j_list)**0.5
RMSE_kf_case4_total = (RMSE_kf_case4_xji + RMSE_kf_case4_yji + RMSE_kf_case4_thetaji + RMSE_kf_case4_vji + RMSE_kf_case4_wji) / 5
print('The x,y,theta, vj, wj RMSE of case 4 :', RMSE_kf_case4_xji, RMSE_kf_case4_yji, RMSE_kf_case4_thetaji, RMSE_kf_case4_vji, RMSE_kf_case4_wji)


cases = ['case1', 'case2', 'case3', 'case4']

values_xji = [ RMSE_kf_case1_xji, RMSE_kf_case2_xji, RMSE_kf_case3_xji, RMSE_kf_case4_xji]
values_yji = [ RMSE_kf_case1_yji, RMSE_kf_case2_yji, RMSE_kf_case3_yji, RMSE_kf_case4_yji]
values_thetaji = [ RMSE_kf_case1_thetaji, RMSE_kf_case2_thetaji, RMSE_kf_case3_thetaji, RMSE_kf_case4_thetaji]


fig = plt.figure(figsize=(5, 7))
fig.subplots_adjust(hspace=0.5)
fig.tight_layout

ax1 = fig.add_subplot(3, 1, 1)
ax1.set_title('$x_{ji}$')
ax2 = fig.add_subplot(3, 1, 2)
ax2.set_title('$y_{ji}$')
ax3 = fig.add_subplot(3, 1, 3)
ax3.set_title('$\Theta_{ji}$')

ax1.bar(cases, values_xji)
ax2.bar(cases, values_yji)
ax3.bar(cases, values_thetaji)

plt.savefig('kf_rmse')
```
